```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import label_binarize
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import roc_auc_score
```

```python
# 获取总数据
# train
train_path = 'data/train_format1.csv'
train_dataset = pd.read_csv(train_path,low_memory=False)
print('train_dataset',train_dataset.shape)
# test
test_path = 'data/test_format1.csv'
test_dataset = pd.read_csv(test_path,low_memory=False)
print('test_dataset',test_dataset.shape)
# all_features
train_features_path = 'data/train_dataset_total.csv'
train_features_dataset = pd.read_csv(train_features_path,low_memory=False)
print('train_features_dataset',train_features_dataset.shape)
train_features_dataset.isnull().any()
```

```python
train_features_dataset.drop(columns=['r1','r2','r3','r4','r5','r6'],inplace=True)
train_dataset = train_dataset.rename(index=str,columns={'merchant_id':'seller_id'})
test_dataset = test_dataset.rename(index=str,columns={'merchant_id':'seller_id'})
```

```python
dataset = train_features_dataset.iloc[:,1:]
```

```python
t_dataset = pd.merge(dataset,train_dataset,on=['user_id','seller_id'],how='right')
t_dataset.drop(columns=['user_id','seller_id'],inplace=True)
print('t_dataset',t_dataset.shape)
t_dataset.isnull().any()
```

```python
# 分割数据
## 获取训练数据
#先进性自测试 
### X_train
X = t_dataset.iloc[:,0:-1]
Y = t_dataset.iloc[:,-1]
### Y_train
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)

#即需要最优化的参数的取值
param_test = {'min_samples_leaf':list(range(1,101,10)), # 40
              'max_depth':range(3,14,2),
              }
# url =>  http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html
gsearch_find_best_params = GridSearchCV(estimator=GradientBoostingClassifier(
                                                                loss = 'deviance', # {‘deviance’ 偏差, ‘exponential’ 指数提升偏差}, optional (default=’deviance’)
                                                               learning_rate=0.1,#  float, optional (default=0.1)
                                                               n_estimators=100, 
                                                               subsample=0.8,
                                                               max_features='sqrt',
                                                               max_depth=8,
                                                                verbose=1,
                                                               min_samples_split=130,
                                                               min_samples_leaf=40,
                                                               # min_weight_fraction_leaf=，
                                                               random_state=10,
                                                               ), #所使用的分类器
                        param_grid=param_test,#即需要最优化的参数的取值，param_grid =param_test1，param_test1 = {'n_estimators':range(10,71,10)}。
                        scoring='roc_auc',#准确度评价标准，默认None,这时需要使用score函数；或者如scoring='roc_auc'，根据所选模型不同，评价准则不同。字符串（函数名），或是可调用对象，需要其函数签名形如：scorer(estimator, X, y)；如果是None，则使用estimator的误差估计函数。
                        refit = True,#:默认为True,程序将会以交叉验证训练集得到的最佳参数，重新对所有可用的训练集与开发集进行，作为最终用于性能评估的最佳模型参数。即在搜索参数结束后，用最佳参数结果再次fit一遍全部数据集。              
                        iid=False, #默认True,为True时，默认为各个样本fold概率分布一致，误差估计为所有样本之和，而非各个fold的平均。
                         verbose=0,#日志冗长度，int：冗长度; 0：不输出训练过程; 1：偶尔输出; >1：对每个子模型都输出。
                        n_jobs = -1,#n_jobs，并行数，int：个数,-1：跟CPU核数一致, 1:默认值。
                        cv=5 #交叉验证参数，默认None，使用三折交叉验证。指定fold数量，默认为3，也可以是yield训练/测试数据的生成器。
)
gsearch_find_best_params.fit(X_train,Y_train)
print(gsearch_find_best_params.cv_results_)
print(gsearch_find_best_params.best_params_, gsearch_find_best_params.best_score_)
print(roc_auc_score(Y_test,gsearch_find_best_params.predict(X_test)))
```

```python
## 获取测试数据
challenge_dataset = pd.merge(dataset,test_dataset,on=['user_id','seller_id'],how='right')
### X_challenge
challenge_dataset.drop(columns=['user_id','seller_id'],inplace=True)
X_challenge = challenge_dataset.iloc[:,0:-1]
### y_predict
X_challenge.count()
```

```python
####for train
gbm = GradientBoostingClassifier(max_features='sqrt',
                                  learning_rate=0.1,
                                  min_samples_leaf=20,
                                  subsample=0.8,
                                  random_state=10,
                                  n_estimators=100)
gbm.fit(X,Y)
y_predprob = gbm.predict(X)
print(roc_auc_score(Y,y_predprob))
# type(y_predprob)
# type(Y_test)
```

```python
y_predict = gbm.predict(X_challenge)
```

```python
import pandas as pd
import numpy as np
```

```python
user_log_path = 'data/user_log_format1.csv'
user_log_dataset = pd.read_csv(user_log_path,low_memory=False)
print(user_log_dataset.head())
```

```python
# u1  u1 用户操作次数 user_handle_day_count
u1 = user_log_dataset[['user_id']].copy()
u1['user_handle_action_count'] = 1
u1 = u1.groupby(['user_id']).agg('sum').reset_index()
u1.head()
```

```python
# u2  用户性别
user_info_path = 'data/user_info_format1.csv'
user_info_dataset = pd.read_csv(user_info_path,low_memory=False)
print(user_info_dataset.head())
```

```python
from sklearn.preprocessing import label_binarize
# 性别 哑编码
u2 = user_info_dataset[['user_id','gender']].copy()
u2 = u2.fillna(2.0)
gender_hot = label_binarize(np.array(u2.gender), classes=[0, 1, 2])
gender_hot_df = pd.DataFrame(gender_hot,columns=['gender_hot_0','gender_hot_1','gender_hot_2'])
u2['gender_hot_0'] = gender_hot_df.gender_hot_0
u2['gender_hot_1'] = gender_hot_df.gender_hot_1
u2['gender_hot_2'] = gender_hot_df.gender_hot_2
u2 = u2.drop(columns=['gender'])
u2.head()
```

```python
#  u3 user_action_0_count user_action_1_count user_action_2_count user_action_3_count
# action 哑编码
u3 = user_log_dataset[['user_id','action_type']].copy()
action_type_hot = label_binarize(np.array(u3.action_type), classes=[0, 1, 2, 3])
action_type_hot_df = pd.DataFrame(action_type_hot,columns=['action_type_hot_0','action_type_hot_1','action_type_hot_2','action_type_hot_3'])
u3['action_type_hot_0'] = action_type_hot_df.action_type_hot_0
u3['action_type_hot_1'] = action_type_hot_df.action_type_hot_1
u3['action_type_hot_2'] = action_type_hot_df.action_type_hot_2
u3['action_type_hot_3'] = action_type_hot_df.action_type_hot_3
u4 = u3.copy()
u4 = u4.groupby('user_id').agg('sum').reset_index()
u4 = u4.rename(index=str,columns={"action_type_hot_0":"action_type_hot_0_count","action_type_hot_1":"action_type_hot_1_count","action_type_hot_2":"action_type_hot_2_count","action_type_hot_3":"action_type_hot_3_count",})
u4 = u4.drop(columns=['action_type'])
u4.head()
```

```python
#u5 用户年龄段  user_age_type
u5 = user_info_dataset[['user_id','age_range']].copy()
u5 = u5.fillna(0)
u5 = u5.replace(8,7)
u5 = u5.drop_duplicates()
age_range_hot = label_binarize(np.array(u5.age_range), classes=[0,1, 2, 3, 4, 5, 6, 7])
age_range_hot_df = pd.DataFrame(age_range_hot,columns=['age_range_0','age_range_1','age_range_2','age_range_3','age_range_4','age_range_5','age_range_6','age_range_7'])
u5['age_range_0'] = age_range_hot_df.age_range_0
u5['age_range_1'] = age_range_hot_df.age_range_1
u5['age_range_2'] = age_range_hot_df.age_range_2
u5['age_range_3'] = age_range_hot_df.age_range_3
u5['age_range_4'] = age_range_hot_df.age_range_4
u5['age_range_5'] = age_range_hot_df.age_range_5
u5['age_range_6'] = age_range_hot_df.age_range_6
u5['age_range_7'] = age_range_hot_df.age_range_7
u5 = u5.drop(columns=['age_range'])
u5.head()
```

```python
# 用户操作了几天
u6 = user_log_dataset[['user_id','time_stamp']].copy()
u6 = u6.drop_duplicates(['user_id','time_stamp'])
u6['user_handle_day_count'] = 1
u6 = u6.groupby(['user_id']).agg('sum').reset_index()
u6.drop(columns=['time_stamp'],inplace=True)
u6.head()
```

```python
# i1 该 cat_id 下面有多少item
i1 = user_log_dataset[['cat_id','item_id']].copy()
i1.drop_duplicates(['item_id'],inplace=True)
i1['cat_item_count'] = 1
i1 = i1.groupby(['cat_id']).agg('sum').reset_index()
i1 = i1.drop(columns=['item_id'])
i1.head()
```

```python
# s1 该 seller 有多少 brand seller_brand_count
s1 = user_log_dataset[['seller_id','brand_id']].copy()
s1.drop_duplicates(['brand_id'],inplace=True)
s1['seller_brand_count'] = 1
s1 = s1.groupby(['seller_id']).agg('sum').reset_index()
s1 = s1.drop(columns=['brand_id'])
s1.head()
```

```python
dataset = total_train_dataset.drop(columns=['user_id','merchant_id'])
dataset.columns
```

```python
X = dataset.iloc[:,0:-1]
Y = dataset.iloc[:,-1]
X.head()
```

```python
Y.head()
Y.count()
```

```python
####for params test
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import roc_auc_score
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
param_test1 = {'min_samples_leaf':list(range(50,81,10))}
gsearch1 = GridSearchCV(estimator=GradientBoostingClassifier(
                                                               learning_rate=0.1,
                                                               n_estimators=100,
                                                               subsample=0.8,
                                                               max_features='sqrt',
                                                               max_depth=8,
                                                               min_samples_split=130,
                                                               verbose=1,
                                                               # min_samples_leaf=30,
                                                               # min_weight_fraction_leaf=，
                                                               random_state=10,
                                                               ),
                        param_grid=param_test1,
                        scoring='roc_auc',
                        iid=False,
                        cv=5
)
gsearch1.fit(X_train,Y_train)
print(gsearch1.cv_results_)
print(gsearch1.best_params_, gsearch1.best_score_)
print(roc_auc_score(Y_test,gsearch1.predict_proba(X_test)[:,1]))
```
